```python
df.drop(['ID','Year_Built'],axis=1,inplace=True)
```

```python
for i in cat:
  print(i,df[i].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in cat:
  df[i]=le.fit_transform(df[i])
```

#####For Classification


```python
median_price = df["Price_in_Lakhs"].median()
df["Good_Investment"] = (df["Price_in_Lakhs"] < median_price).astype(int)
```

```python
x1 = df.drop(["Good_Investment","Price_in_Lakhs"], axis=1)
y1=df["Good_Investment"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y1,random_state=42,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc2=MinMaxScaler()
x_train=sc2.fit_transform(x_train)
x_test=sc2.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
print(lg.score(x_train,y_train))
print(lg.score(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
y_pred = lg.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
```

##### For Regression

```python
x_reg = df.drop('Price_in_Lakhs', axis=1)
y_reg = df['Price_in_Lakhs']
```

```python
x_train_reg, x_test_reg, y_train_reg, y_test_reg = train_test_split(x_reg, y_reg, random_state=42, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc_reg = StandardScaler()
x_train_sc_reg = sc_reg.fit_transform(x_train_reg)
x_test_sc_reg = sc_reg.transform(x_test_reg)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression,Ridge, Lasso
lr_model = LinearRegression()
lr_model.fit(x_train_sc_reg, y_train_reg)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred = lr_model.predict(x_test_sc_reg)
print("MSE: ",mean_squared_error(y_test_reg, y_pred))
print("R2: ",r2_score(y_test_reg, y_pred))
```

```python
# Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(x_train_sc_reg, y_train_reg)
print("Ridge: ",ridge.score(x_test_sc_reg, y_test_reg))

# Lasso
lasso = Lasso(alpha=0.001)
lasso.fit(x_train_sc_reg, y_train_reg)
print('Lasso: ',lasso.score(x_test_sc_reg, y_test_reg))
```
